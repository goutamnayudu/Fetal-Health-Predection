```python
#1
def handle_outliers(x, column):
    Q1 = x[column].quantile(0.25)
    Q3 = x[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    x[column] = x[column].apply(lambda l: x[column].median() if l < lower_bound or l > upper_bound else l)
    return x
for column in x.columns:
  x=handle_outliers(x,column)


x.plot(kind='box',figsize=(17,17),layout=(4,4),sharex=False,subplots=True)
```

```python
for column in ['accelerations','percentage_of_time_with_abnormal_long_term_variability','mean_value_of_long_term_variability','histogram_mode','histogram_variance']:
  x=handle_outliers(x,column)
```

```python
from tensorflow.keras.utils import to_categorical


X_train = x_train_smote
Y_train = y_train_smote
y_test_copy=y_test

print("X_train shape:", X_train.shape)
print("X_test shape:", x_test.shape)
print("Y_train shape:", Y_train.shape)
print("y_test shape:", y_test_copy.shape)
```

```python
Y_train.value_counts()
```

```python
y_test_new = [int(label)-1 for label in y_test_copy]
Y_train_new = [ int(label)-1   for label in Y_train ]


Y_train_new = np.array(Y_train_new)
y_test_new = np.array(y_test_new)


y_train.shape
```

```python
num_classes=3
Y_train=to_categorical(Y_train_new, num_classes)
y_test_cnn=to_categorical(y_test_new, num_classes)
Y_train[:3]
```

```python
Y_train_new.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, GRU, Dense , Flatten
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GRU, Dense, Flatten , Reshape
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.optimizers import Adam
from keras.regularizers import l2
from keras.regularizers import l1
# Assuming the input data has the shape (num_samples, num_timesteps, num_features)
# In your case, num_timesteps will be 1 as you have 5 features for each sample.


num_timesteps = X_train.shape[1]




# Assuming the input data has the shape (num_samples, num_timesteps, num_features)
# In your case, num_timesteps will be 1 as you have 5 features for each sample.

models = Sequential([
    Conv1D(32, 3,  activation=LeakyReLU(alpha=0.01)   , kernel_regularizer=l1(0.01) , padding='same',     input_shape=(8,1)),  # (num_samples, num_timesteps=1, num_features=5)
    MaxPooling1D(2, strides=2),
    Conv1D(64, 3, activation=LeakyReLU(alpha=0.01), padding='same'),
    MaxPooling1D(2, strides=2),
    Flatten(),  # Flatten the output of the Conv1D layers
       Reshape((-1, 64)),
    GRU(64  , dropout=0.4),
    Dense(64, activation=LeakyReLU(alpha=0.01)),
    Dense(3, activation='softmax')  # Output layer with 3 target labels
])


models.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])


models.summary()
```

```python
# Train the model
# batch_size = 64
# epochs = 20

statics=models.fit(X_train, Y_train, batch_size=64, epochs=20 , validation_split=0.2)
```

```python
# Evaluate the model on the test set
loss, accuracy_cnn = models.evaluate(x_test, y_test_cnn)
print("Test Loss:", loss)
print("Test Accuracy:", accuracy_cnn)
```

```python
y_pred_cnn=models.predict(x_test)
y_pred_cnn
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

y_pred_knn = model_knn.predict(x_test)
y_pred_dt = model_dt.predict(x_test)
y_pred_rf = model_rf.predict(x_test)
y_pred_gb = gboost_model.predict(x_test)
y_pred_svc = model_svc.predict(x_test)
y_pred_gscv = model_gscv.predict(x_test)

y_pred_cnn_classes = np.argmax(y_pred_cnn, axis=1) + 1

ensemble_predictions = pd.DataFrame({
    'KNN': y_pred_knn,
    'Decision_Tree': y_pred_dt,
    'Random_Forest': y_pred_rf,
    'Gradient_Boosting': y_pred_gb,
    'SVC': y_pred_svc,
    'GridSearchCV': y_pred_gscv,
    'CNN': y_pred_cnn_classes
})

final_predictions = ensemble_predictions.mode(axis=1)[0]
final_predictions = final_predictions.astype(int)

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print("Testing Accuracy (Max Voting):", accuracy_score(y_test, final_predictions))
print("\nClassification Report:\n", classification_report(y_test, final_predictions))

cm = confusion_matrix(y_test, final_predictions)
cm_display = ConfusionMatrixDisplay(cm).plot()
plt.show()

model_name = np.array(["KNN", "Decision Tree Classifier", "Random Forest Classifier",
                       "Gradient Boosting Classifier", "Support Vector Classifier",
                       "GridSearchCV (Hyperparameter Tuning)", "CNN", "Max Voting Ensemble"])
accuracy_score_of_model = np.array([
    accuracy_score(y_test, y_pred_knn),
    accuracy_score(y_test, y_pred_dt),
    accuracy_score(y_test, y_pred_rf),
    accuracy_score(y_test, y_pred_gb),
    accuracy_score(y_test, y_pred_svc),
    accuracy_score(y_test, y_pred_gscv),
    accuracy_cnn,
    accuracy_score(y_test, final_predictions)
])

performance = pd.DataFrame()
performance['Model Name'] = model_name
performance['Accuracy Score'] = accuracy_score_of_model

print(performance)

plt.figure(figsize=(8, 8))
sns.set(style="whitegrid")
sns.barplot(x='Accuracy Score', y='Model Name', data=performance)
plt.title('Performance Comparison of Models (Including Max Voting)')
plt.show()
```
